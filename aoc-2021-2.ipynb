```python
results = {'horizontal': 0, 'vertical': 0}

def parse_directions(direction):
    parsed_directions = direction.split(' ')
    axis = 'horizontal' if parsed_directions[0] == 'forward' else 'vertical'
    value = int(parsed_directions[1])
    results[axis] += value if parsed_directions[0] != 'up' else value * -1

for direction in input.split('\n'):
    parse_directions(direction)

print(results['horizontal'] * results['vertical'])
```

## Part 2

```python
results = {'horizontal': 0, 'depth': 0}
aim = {'theta': 0}


def translate(x):
    x = x
    results['horizontal'] += x
    results['depth'] += x * aim['theta']


def rotate(x):
    aim['theta'] += x


def calculate_position():
    operations = {'up': lambda x: rotate(x * -1),
                  'down': lambda x:  rotate(x),
                  'forward': lambda x: translate(x)}

    for direction in input.split('\n'):
        split_directions = direction.split(' ')
        operations[split_directions[0]](int(split_directions[1]))

    return results['horizontal'] * results['depth']


print(calculate_position())
```
